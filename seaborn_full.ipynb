```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
df = pd.DataFrame({
    "category": ["A", "A", "B", "B", "B", "C", "C", "C", "C"],
    "value": [1, 2, 2, 3, 1, 4, 4.5, 3.5, 4.2]
})

plt.figure(figsize=(7, 5))

# Strip plot
plt.subplot(1, 2, 1)
sns.stripplot(x="category", y="value", data=df)
plt.title("Strip Plot")

# Swarm plot
plt.subplot(1, 2, 2)
sns.swarmplot(x="category", y="value", data=df)
plt.title("Swarm Plot")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate a dense dataset
np.random.seed(42)
categories = ["A", "B", "C"]
data = {
    "category": np.random.choice(categories, size=300),
    "value": np.random.normal(loc=5, scale=1, size=300)  # Normally distributed data
}
df = pd.DataFrame(data)

# Create the plots
plt.figure(figsize=(7, 5))

# Strip plot
plt.subplot(1, 2, 1)
sns.stripplot(x="category", y="value", data=df, jitter=True, color='red')
sns.stripplot(x="category", y="value", data=df, jitter=False, color='black')
plt.title("Strip Plot jitter (red on) (black off)")


# Swarm plot
plt.subplot(1, 2, 2)
sns.swarmplot(x="category", y="value", data=df)
plt.title("Swarm Plot")

plt.tight_layout()
plt.show()
```

<h3>Perceptually Uniform Sequential</h3>
<p>viridis, plasma, inferno, magma, cividis (great for continuous data and default for many visualizations).</p>

<h3>Sequential</h3>
<p>Blues, Greens, Oranges, Purples, Reds, YlOrBr, YlOrRd (ideal for data progressing from low to high values).</p>

<h3>Diverging</h3>
<p>coolwarm, RdBu, RdYlBu, Spectral, PiYG (useful for data with a meaningful midpoint, such as deviations).</p>

<h3>Cyclic</h3>
<p>twilight, hsv (best for cyclical data like time or angles).</p>

<h3>Qualitative</h3>
<p>Set1, Set2, Paired, tab10 (designed for categorical data with distinct groups).</p>

<h3>Miscellaneous</h3>
<p>turbo, rainbow, terrain, cubehelix (unique colormaps for special cases).</p>

<h1>REGRSSION PLOTS</h1>

```python
# Load dataset
tips_df = sns.load_dataset('tips')

# Set context
sns.set_context('paper', font_scale=1.3)

# Create the lmplot
sns.lmplot(
    x='total_bill', 
    y='tip', 
    hue='sex', 
    data=tips_df, 
    markers=['o', '*'],
    scatter_kws={'s': 100, 'linewidths': 0.5, 'edgecolor': 'b'},  # Fixed the conflict
    height=4,  # Controls the figure size
    aspect=1   # Aspect ratio of the plot
)

# Show the plot
plt.show()
```

```python
sns.set_context('paper', font_scale=1.3)
sns.lmplot(x='total_bill', 
           y='tip', 
           col='sex', 
           row='time', 
           data=tips_df,
           height=4,
           aspect=1)
```
